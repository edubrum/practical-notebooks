<h2>1 N-Queens</h2>

<p><a href="https://en.wikipedia.org/wiki/Eight_queens_puzzle">N-Queens Problem</a> is the problem of placing <i>n</i> chess queens on an chessboard so that no two queens threaten each other; thus, a solution requires that no two queens share the same row, column, or diagonal. The problem is to put <i>n</i> non-attacking queens on an <i>n×n</i> chessboard, for which solutions exist for all natural numbers <i>n</i> with the exception of n = 2 and n = 3. An example of a solution is shown below in Figure 1.</p>


Iterative improvement:
<li>Start with one queen in each column;</li>
<li>Move a queen to reduce number of conflicts.</li>

Even for very large <i>n</i> (e.g., n = 1 million), this usually finds a solution almost instantly.
